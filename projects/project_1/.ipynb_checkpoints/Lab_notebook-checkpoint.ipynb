### 5) Aligning sequences to reference

We will use Burrows-Wheeler Transform for aligning our sequences.

```python
!bwa index
```

```python
%ls
```

```python
!bwa index GCA_000005845.2_ASM584v2_genomic.fna.gz
```

```python
!bwa mem GCA_000005845.2_ASM584v2_genomic.fna.gz output_forward_paired_30.fastq output_reverse_paired_30.fastq > alignment.sam
```

```python
!samtools view -S -b alignment.sam > alignment.bam
```

```python
!samtools flagstat alignment.bam
```

```python
!samtools sort alignment.bam -o alignment_sorted.bam
```

```python
!samtools index alignment_sorted.bam
```
